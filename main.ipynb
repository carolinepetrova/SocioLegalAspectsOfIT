```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)


def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
```

```python
universities_grouped = universities.groupby(['Професионално направление']).sum().reset_index()
universities_grouped_sum = universities_grouped.sum()

x = universities_grouped['Професионално направление']
y = universities_grouped['Общ брой']
x_pos = [i for i, _ in enumerate(x)]

fig, ax = plt.subplots()
rects1 = ax.bar(x_pos, y, color='coral')
plt.xlabel("Висши училища")
plt.ylabel("Брой на студенти")
plt.title("Брой на студентите по професионално направление")
plt.xticks(x_pos, x,rotation='vertical')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2, 1*height,
                 int(height),ha='center', va='bottom')
autolabel(rects1)
plt.show()
```
